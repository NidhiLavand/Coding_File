```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data=pd.read_csv("/content/train.csv")
test_data=pd.read_csv("/content/test.csv")
```

```python
train_data.info()
```

```python
sur=sum(train_data["Survived"]==0)
```

```python
not_sur=sum(train_data["Survived"]==1)
```

```python
category=["Survived", "Non-servived"]
Count=[sur,not_sur]
plt.bar(category,Count)
plt.title("Information on survival rate")
plt.xlabel("Category")
plt.ylabel("Number")
plt.show()
```

```python
train_data.drop(columns=["PassengerId","Name","Ticket","Pclass","Age","Fare","Cabin"],inplace=True)
test_data.drop(columns=["PassengerId","Name","Ticket","Pclass","Age","Fare","Cabin"],inplace=True)
```
